```python
class rockpaperscissor():        #creating an object with name rockpaperscissor.
    userscore = 0                #Defining some attributes.
    computerscore = 0
    numofties = 0
    def __init__(self, userscore, computerscore, numofties):
        self.userscore = userscore
        self.computerscore = computerscore
        self.numofties = numofties
        print('Rock Paper Scissor object has been created')
    def rps(self):               #creating a method for the game rock paper scissor.
        import random            #importing random to generate random choice for computer.
        lst = ['rock','paper','scissor']   #creating a list with strings i.e., rock paper and scissor.
        print('\n1.Rock\n2.Paper\n3.Scissor\n4.Exit\n')   #printing the options for the user to decide among them.
        print('\nRock Paper Scissor ?')
        while lst:      
            userchoice = input('\nEnter your choice : ')     #asking for user choice.
            if userchoice != 'exit':                         #checks for this condition,to generate random choice for computer.
                computerchoice = random.choice(lst)
                print('\n------------------------------')
                print('\nComputer choice is :',computerchoice)
            if userchoice == 'rock':
                print('\nUser choice is :',userchoice)      #prints userchoice as rock if the condition satisfies.
                print('\n------------------------------')
            elif userchoice == 'paper':
                print('\nUser choice is :',userchoice)    #prints userchoice as paper if the condition satisfies.
                print('\n------------------------------')
            elif userchoice =='scissor':    
                print('\nUser choice is :',userchoice)    #prints userchoice as scissor if the condition satisfies.
                print('\n------------------------------')
            elif userchoice == 'exit':
                print('\nUser wants to exit the game')     #prints this msg when the user wants to exit the game.
                print('\n------------------------------')
                break
            else:
                print('\nYour choice is invalid')   #prints this msg if the user chooses other than the mentioned options.
                print('\n------------------------------')
                break
            while userchoice != 'exit':          #creating this loop to check for the winner between user and computer.
                if userchoice == 'rock' and computerchoice == 'scissor':
                    print('\nRock smashes scissor')    #prints when the above condition satisfies.
                    print('\nUser wins')
                    self.userscore = self.userscore + 1      #for updating the score.
                    print('\n----------Score Card----------')
                    print('\nUserscore =',self.userscore)
                    print('\nComputerscore =',self.computerscore)
                    print('\nNumber of ties =',self.numofties)                
                    print('\n------------------------------')
                    break
                elif userchoice == 'scissor' and computerchoice == 'rock':
                    print('\nRock smashes scissor')     #prints when the above condition satisfies.
                    print('\nComputer wins')
                    self.computerscore = self.computerscore + 1    #for updating the score.
                    print('\n----------Score Card----------')
                    print('\nUserscore =',self.userscore)
                    print('\nComputerscore =',self.computerscore)
                    print('\nNumber of ties =',self.numofties)
                    print('\n------------------------------')
                    break
                elif userchoice == 'rock' and computerchoice == 'paper':
                    print('\nPaper catches rock')      #prints when the above condition satisfies.
                    print('\nComputer wins')
                    self.computerscore = self.computerscore + 1   #for updating the score.
                    print('\n----------Score Card----------')
                    print('\nUserscore =',self.userscore)
                    print('\nComputerscore =',self.computerscore)                    
                    print('\nNumber of ties =',self.numofties)
                    print('\n------------------------------')
                    break
                elif userchoice == 'paper' and computerchoice == 'rock':
                    print('\nPaper catches rock')       #prints when the above condition satisfies.
                    print('\nUser wins')
                    self.userscore = self.userscore + 1    #for updating the score.
                    print('\n----------Score Card----------')
                    print('\nUserscore =',self.userscore)
                    print('\nComputerscore =',self.computerscore)
                    print('\nNumber of ties =',self.numofties)
                    print('\n------------------------------')
                    break
                elif userchoice == 'paper' and computerchoice == 'scissor':
                    print('\nScissor cuts paper')    #prints when the above condition satisfies.
                    print('\nComputer wins')
                    self.computerscore = self.computerscore + 1      #for updating the score.
                    print('\n----------Score Card----------')
                    print('\nUserscore =',self.userscore)
                    print('\nComputerscore =',self.computerscore)
                    print('\nNumber of ties =',self.numofties)
                    print('\n------------------------------')
                    break
                elif userchoice == 'scissor' and computerchoice == 'paper':
                    print('\nScissor cuts paper')      #prints when the above condition satisfies.
                    print('\nUser wins')
                    self.userscore = self.userscore + 1     #for updating the score.
                    print('\n----------Score Card----------')
                    print('\nUserscore =',self.userscore)
                    print('\nComputerscore =',self.computerscore)
                    print('\nNumber of ties =',self.numofties)
                    print('\n------------------------------')
                    break
                elif userchoice == 'rock' and computerchoice == 'rock':
                    print('\nUserchoice and computer choice both are same')     #prints when the above condition satisfies.
                    print('\nHence its a tie')
                    self.numofties = self.numofties + 1     #for updating the score.
                    print('\n----------Score Card----------')
                    print('\nUserscore =',self.userscore)
                    print('\nComputerscore =',self.computerscore)
                    print('\nNumber of ties =',self.numofties)
                    print('\n------------------------------')
                    break
                elif userchoice == 'paper' and computerchoice == 'paper':
                    print('\nUserchoice and computer choice both are same')     #prints when the above condition satisfies.
                    print('\nIts a tie')
                    self.numofties = self.numofties + 1     #for updating the score.
                    print('\n----------Score Card----------')
                    print('\nUserscore =',self.userscore)
                    print('\nComputerscore =',self.computerscore)
                    print('\nNumber of ties =',self.numofties)
                    print('\n------------------------------')
                    break
                elif userchoice == 'scissor' and computerchoice == 'scissor':
                    print('\nUserchoice and computer choice both are same')     #prints when the above condition satisfies.
                    print('\nIts a tie')
                    self.numofties = self.numofties + 1      #for updating the score.
                    print('\n----------Score Card----------')
                    print('\nUserscore =',self.userscore)
                    print('\nComputerscore =',self.computerscore)
                    print('\nNumber of ties =',self.numofties)
                    print('\n------------------------------')
                    break
                else:
                    print('\nSince your entered choice is invalid, please try again') #prints this msg if the choice is invalid.
                    break
```
